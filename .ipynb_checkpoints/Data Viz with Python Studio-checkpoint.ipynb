```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", error_bad_lines=False)

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years
```

With your dataframe set up, time to start creating visualizations!

```python
df.info()
```

```python
df['year'].describe()
```

```python
# Use your Python skills to create a visualization here.
df_year = df.sort_values(by=['year'], ascending=True)
# library & dataset
import seaborn as sns
# Plot the histogram thanks to the distplot function
fig= sns.histplot(df_year["year"], color = 'seagreen', kde=True)
for ind, label in enumerate(fig.get_xticklabels()):
    if ind % 10 == 0:  # every 15th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)
plt.title('Number of Books Published per Year', fontweight='bold')
plt.xlabel('years', fontweight='bold')
plt.ylabel('number of books', fontweight ='bold')
```
